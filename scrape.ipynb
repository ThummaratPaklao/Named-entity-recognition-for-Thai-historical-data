```python
def wikibot(url):
  cont=[]
  title=[]
  url_open =requests.get(url)
  soup = BeautifulSoup(url_open.content, 'html.parser')
  details = soup('table',{'class':'infobox'})
  for i in range(1,len(soup('p'))):
    #print(soup('p')[i].text)
    cont.append(soup('p')[i].text)
    title.append(soup.title.string)
  return cont,title
```

```python
cont1=[]
url_open =requests.get("https://th.wikipedia.org/wiki/%E0%B8%81%E0%B8%9A%E0%B8%8F%E0%B8%AB%E0%B8%A7%E0%B8%B1%E0%B8%99%E0%B8%AB%E0%B8%A1%E0%B8%B2%E0%B8%94%E0%B8%AB%E0%B8%A5%E0%B8%B5")
urls = ['https://th.wikipedia.org/wiki/%E0%B8%81%E0%B8%9A%E0%B8%8F%E0%B8%AB%E0%B8%A7%E0%B8%B1%E0%B8%99%E0%B8%AB%E0%B8%A1%E0%B8%B2%E0%B8%94%E0%B8%AB%E0%B8%A5%E0%B8%B5']
soup = BeautifulSoup(url_open.content, 'html.parser')
title = []
for i in range(1,len(soup('p'))):
  #print(soup('p')[i].text)
  cont1.append(soup('p')[i].text)
  title.append(soup.title.string)

for i in range(1,len(soup('p'))):
  for link in soup('p')[i].find_all('a'):
    if link != []:
      print(link.get('href'))
      cc = "https://th.wikipedia.org"+str(link.get('href'))
      if cc not in urls:
        urls.append(cc)
        c = wikibot(cc)
        for j in c[0]:
          cont1.append(j)
        for j in c[1]:
          title.append(j)
        print(j)
```
