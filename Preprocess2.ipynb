```python
import pandas as pd
import numpy as np
import re
```

```python
title_content = pd.read_csv('/content/drive/MyDrive/Intern_PSU_Phuket/Scrape_Wiki/title_content_newadd.csv')
```

```python
title_content
```

```python
row_repeat = []
title_uni = []
for index, row in title_content.iterrows():
  if index == 0:
    row_repeat.append(0)
    title_uni.append(row['title'])
  elif index > 0:
    if row['title'] == title_content['title'][index-1]:
      row_repeat.append(row_repeat[index-1])
    elif row['title'] != title_content['title'][index-1] and row['title'] not in title_uni:
      row_repeat.append(0) 
      title_uni.append(row['title'])
    elif row['title'] != title_content['title'][index-1] and row['title'] in title_uni:
      row_repeat.append(1) 
```

```python
index_repeat=[]
for i in range(len(row_repeat)):
  if row_repeat[i]==1:
    index_repeat.append(i)
```

```python
data_new = title_content.drop(index_repeat, axis=0)
```

```python
data_new = data_new.reset_index(drop=True)
```

```python
data_new
```

```python
index_rem = []
for index, row in data_new.iterrows():
  pattern="\d+"
  str_list=re.findall(pattern, row['title'])
  if 'Not Found' in row['title']:
    index_rem.append(index)
  elif 'กำลังสร้าง' in row['title']:
    index_rem.append(index)
  elif 'สารานุกรมเสรี' in row['title']:
    index_rem.append(index)
  elif str_list != []:  
    index_rem.append(index)  
```

```python
data_new1 = data_new.drop(index_rem, axis=0)
data_new1 = data_new1.reset_index(drop=True)
```

```python
data_new1
```

```python
index_rem = []
for index, row in data_new1.iterrows():
  a = str(data_new1['content'][index]).split( )
  if 'เนื้อหา' == row['content']:
    index_rem.append(index)
  elif '\n' == row['content'] or '' == row['content']:
    index_rem.append(index)
  elif '°' in str(row['content']):
    index_rem.append(index)  
  elif '%' in str(row['content']):
    index_rem.append(index)  
  elif '=' in str(row['content']):
    index_rem.append(index)
  data_new1['content'][index] = re.sub("[\<>\[]*?[\]\;]", "", str(row['content']))
  data_new1['content'][index] = str(data_new1['content'][index]).replace('(*)','')
  data_new1['content'][index] = str(data_new1['content'][index]).replace(':','') 
  data_new1['content'][index] = str(data_new1['content'][index]).replace('\n','')
```

```python
data_new2 = data_new1.drop(index_rem, axis=0)
data_new2 = data_new2.reset_index(drop=True)
```
