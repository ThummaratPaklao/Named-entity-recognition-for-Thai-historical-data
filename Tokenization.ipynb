```python
len(old_file)
```

```python
t_data=[]
old_index=-1
cont =[]
for ind, row in data.iterrows():
  index = list(data['title'].unique()).index(str(row['title']))
  if index > old_index:
    t_data.append(str(row['content']))
    cont.append(str(row['title']))
    old_index=index
  else:
    t_data[index] = t_data[index] +' ' + str(row['content'])
```

```python
ti_rem=[]
cont_rem=[]
for i in range(len(cont)): 
  a=str(cont[i]).replace(' - วิกิพีเดีย','')
  if a+'.csv' in old_file or '/' in a:
    ti_rem.append(cont[i])
    cont_rem.append(t_data[i])
```

```python
for i in ti_rem:
  cont.remove(i)
for i in cont_rem:
  t_data.remove(i)
```

```python
dat = {'title':cont,'content':t_data}
dat1=pd.DataFrame(dat)
dat1.to_csv('/content/drive/MyDrive/Intern_PSU_Phuket/Scrape_Wiki/pre_title_content_newadd2.csv',index=False)
```

```python
indexx =0 
for text in t_data:
  tokennewmm_text = word_tokenize(text, engine="newmm")
  tokendc_text = word_tokenize(text, engine="deepcut")
  tokenat_text = word_tokenize(text, engine="attacut")
  tok_newmm = pd.DataFrame(tokennewmm_text)
  tok_dc =  pd.DataFrame(tokendc_text)
  tok_at =  pd.DataFrame(tokenat_text)
  tok_newmm.to_csv('/content/drive/MyDrive/Intern_PSU_Phuket/Token_data/newadd/nm/'+str(cont[indexx]).replace(' - วิกิพีเดีย','')+'.csv',index=False,header=False)
  tok_dc.to_csv('/content/drive/MyDrive/Intern_PSU_Phuket/Token_data/newadd/dc/'+str(cont[indexx]).replace(' - วิกิพีเดีย','')+'.csv',index=False,header=False)
  tok_at.to_csv('/content/drive/MyDrive/Intern_PSU_Phuket/Token_data/newadd/at/'+str(cont[indexx]).replace(' - วิกิพีเดีย','')+'.csv',index=False,header=False)
  indexx = indexx+1
```
